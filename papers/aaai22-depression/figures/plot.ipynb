```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = {
    "posts": [16, 32, 64],
    "Depress": [0.709, 0.738, 0.765],
    "BDI-II": [0.697, 0.721, 0.769],
    "Full": [0.701, 0.743, 0.774]
}
df = pd.DataFrame(df)
df
```

```python
plt.rcParams.update({'font.size': 14})
markers = {'BDI-II': '*', 'Depress': 'x', 'Full': 's'}
for label in ['BDI-II', 'Depress', 'Full']:
    plt.plot(df['posts'], df[label], label=label, marker=markers[label])

plt.legend(loc='best')
plt.xlabel("#Posts")
plt.ylabel("F1 Score")
plt.xticks(df['posts'], df['posts'])
plt.tight_layout()
plt.savefig("post_numbers.pdf", dpi=160)
plt.show()
```
