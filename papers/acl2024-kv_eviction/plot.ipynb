```python
import matplotlib.pyplot as plt

# 数据
labels = ['AAS', 'AQAS', 'LTAS', 'MAS']
data = [
    [0.72, 0.73, 0.78, 0.82],  # AAS
    [0.48, 0.49, 0.57, 0.66],  # AQAS
    [0.76, 0.77, 0.81, 0.87],  # LTAS
    [0.95, 0.97, 0.98, 0.99]   # MAS
]

plt.rcParams.update({'font.size': 13})

# x轴的刻度，这里假设是四个时间点
x = [0.3, 0.4, 0.5, 0.6]

# 标记列表，每个标记对应一条折线
markers = ['o', 's', '^', 'D']

# 绘制折线图
for i, (label, d, marker) in enumerate(zip(labels, data, markers)):
    plt.plot(x, d, marker=marker, label=label)

# 添加图例
plt.legend()

# 添加标题和轴标签
# plt.title('Con')
plt.xlabel('KV cache budget rate')
plt.ylabel('Jaccard Similarity')

# 显示网格
plt.grid(True)
# plt.xlim(0.0, 1.0)
plt.ylim(0.2, 1.0)

# 显示图表
plt.savefig("./figures/consistency.pdf", dpi=360)
plt.show()
```

```python
import matplotlib.pyplot as plt

# 数据
fig, axs = plt.subplots(1,2, figsize=(8,4))
labels = ['Local Window', 'Standard Deviation']
rouge_data = [
    [48.0, 45.7, 43.4, 40.2, 38.0],
    [51.1, 49.6, 50.6, 48.1, 51.0],
]

meteor_data = [
    [53, 50.4, 48.5, 43.8, 45.1],
    [56.0, 54, 55.6, 52.6, 56.7]
]

x = [0.125, 0.250, 0.375, 0.50, 0.625]
markers = ['o', 's']

plt.rcParams.update({'font.size': 13})
for i in range(2):
    # 绘制折线图
    data = rouge_data if i == 0 else meteor_data
    for j, (label, d, marker) in enumerate(zip(labels, data, markers)):
        axs[i].plot(x, d, marker=marker, label=label)
    if i==0: axs[i].set_title("Geo mean of ROUGE-1/2/L")
    else: axs[i].set_title("Meteor")
    axs[i].grid(True)
    axs[i].set_xlabel("Proportion w.r.t cache size")
    axs[i].legend(loc='lower left')

# 显示图表
plt.tight_layout()
plt.savefig("./figures/scope.pdf", dpi=360)
plt.show()
```
